```python
import random

types = ('Hearts', 'Diamonds', 'Clover', 'Spades')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two': 2, 'Three': 3,'Four': 4, 'Five': 5,'Six': 6, 'Seven': 7,
          'Eight': 8, 'Nine': 9,'Ten': 10, 'Jack': 10, 'Queen': 10, 'King': 10, 'Ace': 11}
```

```python
class Deck():

    def __init__(self):
        self.deck = []
        
        for r in ranks:
            for t in types:
                a = Card(r,t)
                
                self.deck.append(a)
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal_one(self):
        return self.deck.pop()
```

```python
class Card():
    
    def __init__(self,ranks,types):
        self.ranks = ranks
        self.types = types
        self.value = values[ranks]
    
    def __str__(self):
        return self.ranks + " of " + self.types
```

```python
class Player():
    
    def __init__(self):
        self.players_cards = []
        self.cash = 100
        self.cashback = 0
        self.Bet = 0
        self.bettings = 0
        self.scoring = 0
    
    def bet(self,ammount):
        self.cashback = ammount
        self.cash = self.cash - ammount
        self.bettings = self.bettings + ammount
        self.Bet += self.cashback + ammount
        print(f"Ammount betted: {ammount}. Acumulated player's bet: {self.bettings}. Total bets: {self.Bet}")
        print(f"Player's cash: {self.cash}")
        
    def not_bet(self):
        print("Player opted not to bet")
        
    def stand(self):
        return True
        
    def hit(self):
        return True
    
    def rounds(self,result):
        self.players_cards = []
        
        if result == 1:
            self.cash = self.Bet + self.cash
            self.Bet = 0
            self.cashback = 0
            self.bettings = 0
        
        if result == 2:
            self.Bet = 0
            self.cashback = 0
            self.bettings = 0
        
        if result == 3:
            self.Bet = 0
            self.cashback = 0
            self.bettings = 0
        
        
```

```python
class Dashboard():
    
    def __init__(self):
        self.victory = 0
        self.losses = 0
        self.busts = 0
    
    def show(self,status):
        if status == 1:
            self.victory += 1
        if status == 2:
            self.losses += 1
        if status == 3:
            self.busts += 1
        print(f"Player Wins: {self.victory}. Player Losses: {self.losses}. Player Busts: {self.busts}")
```

```python
class Game():
    
    def __init__(self):
        self.betting = False
        self.actioning = False
        self.roundy = False
        self.gamy = False
        self.number = False
        self.ace = False
```

```python
class automated_dealer():
    
    def __init__(self):
        self.dealer_cards = []
        self.score = 0
    
    def score(self):
        return random.randint(15,26)
```

```python
game = True
Player_One = Player()
dashboard = Dashboard()
gamer = Game()
deck = Deck()
deck.shuffle()
dealers = automated_dealer()
dealers.score = 0

while game == True:
    
    Player_One.players_cards.append(deck.deal_one())
    print(f'Dealer deals a {Player_One.players_cards[-1]} to the Player')
    
    if dealers.score < 16:
        dealers.dealer_cards.append(deck.deal_one())
        print(f'Automated Dealer deals a {dealers.dealer_cards[-1]} to the "Itself"')
        dealers.score = dealers.dealer_cards[-1].value + dealers.score
    
    print("Players Cards: ")
    for i,j in enumerate(Player_One.players_cards):
        print(j)
    
    while gamer.actioning == False:
        try:
            action = int(input("1. Hit | 2. Stand: "))
            if action == 1 or action == 2:
                gamer.actioning = True
        except ValueError:
            print("Please enter a valid number")
    
    if action == 1:
        print('Time for bets!')
        while gamer.betting == False:
            bets = input("Wanna bet? Y or N: ")
            if bets == 'Y':
                while gamer.number == False:
                    try:
                        cash = int(input("Select the ammount to bet: "))
                        Player_One.bet(cash)
                        if cash >= 0:
                            gamer.number = True
                    except ValueError:
                        print("Insert a valid number")
            if bets == 'N':
                Player_One.not_bet()
            if bets == 'Y' or bets == 'N':
                gamer.betting = True
        gamer.roundy = False
    elif action == 2:
        gamer.roundy = True
        while gamer.betting == False:
            bets = input("Wanna bet? Y or N: ")
            if bets == 'Y':
                while gamer.number == False:
                    try:
                        cash = int(input("Select the ammount to bet: "))
                        Player_One.bet(cash)
                        if cash >= 0:
                            gamer.number = True
                    except ValueError:
                        print("Insert a valid number")    
            if bets == 'N':
                Player_One.not_bet()
            if bets == 'Y' or bets == 'N':
                gamer.betting = True
              
        aces = 0
        while dealers.score < 16:
            dealers.dealer_cards.append(deck.deal_one())
            print(f'Automated Dealer deals a {dealers.dealer_cards[-1]} to the "Itself"')
            dealers.score = dealers.dealer_cards[-1].value + dealers.score
            
            if dealers.dealer_cards[-1].value == 11:
                aces = 1
            if dealers.score > 21 and aces == 1:
                dealers.score = dealers.score - 10
                
                
                
            
        for i in Player_One.players_cards:
            gamer.ace == False
            if i.ranks == 'Ace':
                while gamer.ace == False:
                    try:
                        ace_value = int(input("Select a value for the Ace: "))
                        i.value = ace_value
                        if ace_value == 1 or ace_value == 11:
                            gamer.ace = True
                    except ValueError:
                        print("Values must be 1 or 11")
            Player_One.scoring += i.value
            
        print(f'Player Score: {Player_One.scoring}')
        print(f'Automated Dealer Score: {dealers.score}')
        
        win = 0
        if Player_One.scoring > dealers.score and Player_One.scoring <= 21 and dealers.score <= 21:
            dashboard.show(1)
            win = 1
        elif Player_One.scoring < dealers.score and dealers.score > 21 and Player_One.scoring <= 21:
            dashboard.show(1)
            win = 1
        elif Player_One.scoring < dealers.score and dealers.score <= 21:
            dashboard.show(2)
            win = 2
        elif Player_One.scoring > 21:
            dashboard.show(3)
            win = 3
        else:
            continue
            

    while gamer.roundy == True:
        while gamer.gamy == False:
            game_on = input("Wanna play another round? Y or N: ")
            if game_on == 'Y':
                Player_One.scoring = 0
                dealers.score = 0
                deck = Deck()
                deck.shuffle()
                game = True
                print(win)
                if win == 1:
                    Player_One.rounds(1)
                if win == 2:
                    Player_One.rounds(2)
                if win == 3:
                    Player_One.rounds(3)
                
            elif game_on == 'N':
                game = False
            if game_on == 'Y' or game_on == 'N':
                gamer.gamy = True
        gamer.roundy = False
        
    action = 0
    bets = ''
    game_on = ''
    gamer.gamy = False
    gamer.actioning = False
    gamer.betting = False
    gamer.number = False
```
